Boxes: The whole table

Rule: have to fill in all the remaining numbers that has no contricdation and contails all the number from 1 to 9 


# Building the table

```python
#Building the grids: There would be 9 rows with 9 columns A-I and 1-9 accordingly

rows = 'ABCDEFGHI'
columns = '123456789'

#creating a fuction that could form a concatenations of a letter for example with "ABC" and "123" we would have 
# 9 combination: A1, A2, A3, B1, B2, B3, C1, C2, C3

def cross(a,b):
  return [i+j for i in a for j in b]

```

Now we can create the labels of the boxes, units:

```python
boxes = cross(rows, columns)
print(boxes)
```

```python
row_units = [cross(r, columns) for r in rows]

column_units = [cross(rows, c) for c in columns]

square_units = [cross(r, c) for r in ('ABC', 'DEF', 'GHI') for c in ('123', '456', '789')]
#Test
#print(row_units, column_units)
#print(square_units)

unitlist = row_units + column_units + square_units
#Now make it becomes a dictionary:
#where each unit contails by all square, row, colum around it
#for example A1 will be in square that has A1, row has A1 and column has A1

units = dict((s, [u for u in unitlist if s in u]) for s in boxes)
peers = dict((s, set(sum(units[s], [])) - set([s])) for s in boxes)
print(unitlist)
print(peers)
print(units)
```

# Building the Table value

```python
#Building display 2D Soduku

def display(values):
  """
  The values are the sodoku in dictionary form

  """
  
  width = 1+max(len(values[s]) for s in boxes) #width of each element (2 more more spaces)

  #print(width)
  #for seperate
  line = '+'.join(['-'*(width*3)]*3)
  for r in rows:
    print(''.join(values[r+c].center(width)+('|' if c in '36' else '') for c in columns ))
    if r in 'CF': print(line)
  return

#Testing

def grid_values (grid):
  """This fuction will turn grid 
  from a string (len should be 81) 
  into a dictionary
  For example: if the grid string is ..3 it also means A1: ., A2:., A3: 3)
  """
  #make sure it is 81 in length
  values = []
  alldigits = '123456789'
  for c in grid:
    if c == '.':
      values.append(alldigits)
    elif c in alldigits:
      values.append(c) 
  assert len(grid) == 81
  return dict(zip(boxes,values))

#grid = (grid_values('..3.2.6..9..3.5..1..18.64....81.29..7.......8..67.82....26.95..8..2.3..9..5.1.3..'))
grid = grid_values('4.....8.5.3..........7......2.....6.....8.4......1.......6.3.7.5..2.....1.4......')
#print(grid)
display(grid)
#Step one: do eliminations
def eliminates(values):

  solved_values = [box for box in values.keys() if len(values[box]) == 1]

  for box in solved_values:
    digit = values[box]
    for peer in peers[box]:
      values[peer] = values[peer].replace(digit,'')
  return values

#newvalues = eliminates(grid)
#display(newvalues)
```

```python
#step 2: do the only-choice rule
def only_choice(values):
  '''
  The input would be the soduku with dict form
  '''
  for unit in unitlist:
      for digit in '123456789':
          replace = [box for box in unit if digit in values[box]]
          
          #print(replace)
          if len(replace) == 1:
              values[replace[0]] = digit
              
  return values
display(grid)
#testing
#display(only_choice(newvalues))
```

```python
#Constrains

def constrains(values):
  '''
  Continously using eliminate and only choice to solve the soduku

  '''
  complete = False

  while not complete:
    #check how many boxes have a determined value
    solved_values_before = len([box for box in values.keys() if len(values[box]) == 1])
    #step 1: eliminates
    values = eliminates(values)
    #step2: only_choices
    values = only_choice(values)
    display(values)
    #check how many boxes have a determined value
    solved_values_after = len([box for box in values.keys() if len(values[box]) == 1])
    complete = solved_values_after = solved_values_before

    #double check, reutrn false if there is a box iwth zero availables value
    if len([box for box in values.keys() if len(values[box]) == 0]):
      return False
  return values

#Using DFS to finish


def search(values):
    "Using depth-first search and propagation, try all possible values."
    # First, reduce the puzzle using the previous function
    values = constrains(values)
    if values is False:
        return False ## Failed earlier
    if all(len(values[s]) == 1 for s in boxes): 
        return values ## Solved!
    # Choose one of the unfilled squares with the fewest possibilities
    n,s = min((len(values[s]), s) for s in boxes if len(values[s]) > 1)
    # Now use recurrence to solve each one of the resulting sudokus, and 
    for value in values[s]:
        new_sudoku = values.copy()
        new_sudoku[s] = value
        attempt = search(new_sudoku)
        if attempt:
            return attempt
solved = search(grid)
display(solved)
```
